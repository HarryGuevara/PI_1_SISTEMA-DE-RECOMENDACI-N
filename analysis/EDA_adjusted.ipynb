```python
import pandas as pd
import re

# Función para preprocesar el texto
def preprocess_text(text):
    if pd.isna(text):
        return ''
    text = text.lower()  # Convertir a minúsculas
    text = re.sub(r'[^\w\s]', '', text)  # Eliminar puntuación
    
    # Palabras comunes en español que queremos descartar
    stop_words = {'de', 'la', 'y', 'el', 'los', 'en', 'del', 'una', 'un', 'a', 'por', 'para', 'con', 'se'}
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]
    
    return ' '.join(filtered_words)

# Cargar y procesar los datasets
file_paths = {
    'cast_df': 'C:/Users/maron.LAPTOP-UEJMOSD4/Desktop/my_recommendation_project/data/processed/cast_cleaned.csv',
    'crew_df': 'C:/Users/maron.LAPTOP-UEJMOSD4/Desktop/my_recommendation_project/data/processed/crew_cleaned.csv',
    'movie_df': 'C:/Users/maron.LAPTOP-UEJMOSD4/Desktop/my_recommendation_project/data/processed/df_movie_progress (3).csv'
}

# Columnas de texto para preprocesar en cada archivo
text_columns = {
    'movie_df': ['title', 'overview', 'genres', 'tagline'],
    'cast_df': ['name', 'character'],  # Supuestas columnas de texto en cast_df
    'crew_df': ['name', 'job']         # Supuestas columnas de texto en crew_df
}

# Cargar y procesar cada dataset
for df_name, file_path in file_paths.items():
    df = pd.read_csv(file_path)
    
    # Aplicar preprocesamiento a cada columna de texto especificada
    if df_name in text_columns:
        for col in text_columns[df_name]:
            if col in df.columns:
                df[col] = df[col].apply(preprocess_text)
    
    # Mostrar una muestra del DataFrame para confirmar el procesamiento
    print(f"\nMuestra de {df_name}:")
    print(df.head())
```

```python
from collections import Counter
import re

# Lista de palabras conectores en inglés
connectors = {"the", "a", "and", "to", "of", "in", "is", "his", "her", "for", "he", "on", "by", "that", "an", "who", "as", "their", "with", "I", "it", "you", "your", "are", "was", "one"}

# Columnas de texto a analizar en cada DataFrame
text_columns = {
    'movie_df': ['title', 'overview', 'genres', 'tagline'],
    'cast_df': ['character', 'name'],
    'crew_df': ['department', 'job', 'name']
}

# Función para obtener palabras frecuentes en una columna específica, con opción de eliminar conectores
def get_most_common_words(df, column, num_words=20, remove_connectors=False):
    words = ' '.join(df[column].dropna()).split()
    
    # Eliminar conectores solo si remove_connectors es True
    if remove_connectors:
        words = [word for word in words if word.lower() not in connectors]
    
    word_counts = Counter(words)
    return word_counts.most_common(num_words)

# Aplicar la función a cada columna de interés en cada DataFrame
for df_name, columns in text_columns.items():
    current_df = globals()[df_name]  # Obtiene el DataFrame por su nombre en el diccionario
    
    print(f"\nPalabras más comunes en {df_name}:")
    
    for col in columns:
        # Eliminar conectores solo en las columnas 'overview' y 'tagline'
        remove_connectors = col in ['overview', 'tagline']
        if col in current_df.columns:
            common_words = get_most_common_words(current_df, col, remove_connectors=remove_connectors)
            print(f"\nColumna '{col}':")
            for word, count in common_words:
                print(f"{word}: {count}")
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Función para preprocesar el texto
def preprocess_text(text):
    if pd.isna(text):
        return ''
    text = text.lower()  # Convertir a minúsculas
    text = re.sub(r'[^\w\s]', '', text)  # Eliminar puntuación
    return text

# Definir las columnas de texto relevantes para cada DataFrame
text_columns = {
    'movie_df': ['title', 'overview', 'genres', 'tagline'],
    'cast_df': ['character', 'name'],
    'crew_df': ['department', 'job', 'name']
}

# Preprocesar cada columna de texto en cada DataFrame
for df_name, columns in text_columns.items():
    df = globals()[df_name]  # Obtener el DataFrame por su nombre
    for col in columns:
        if col in df.columns:
            df[col] = df[col].apply(preprocess_text)

# Función para obtener texto combinado de varias columnas de un DataFrame
def get_combined_text(df, columns):
    all_texts = {}
    for col in columns:
        if col in df.columns:
            text = ' '.join(df[col].dropna().astype(str))
            if text:
                all_texts[col] = text
            else:
                print(f'No text found in column: {col}')
    combined_text = ' '.join(all_texts.values())
    return combined_text

# Lista de palabras a excluir (stopwords)
stopwords = set(['id', 'movie', 'film', 'data', 'name'])  # Puedes agregar más

# Función para contar la frecuencia de las palabras y mostrar una gráfica
def generate_wordcloud_alt(text, title, stopwords):
    # Filtrar el texto, eliminar las stopwords
    words = text.split()
    filtered_words = [word for word in words if word not in stopwords]
    
    # Contar las frecuencias de las palabras
    word_counts = Counter(filtered_words)
    
    # Obtener las palabras y las frecuencias más comunes
    most_common = word_counts.most_common(50)
    
    # Preparar los datos para la visualización
    words, counts = zip(*most_common)
    
    # Graficar las palabras más comunes usando Matplotlib
    plt.figure(figsize=(10, 6))
    y_pos = np.arange(len(words))
    
    plt.barh(y_pos, counts, align='center', color='skyblue')
    plt.yticks(y_pos, words)
    plt.xlabel('Frecuencia')
    plt.title(title)
    plt.gca().invert_yaxis()  # Invertir el eje Y para mostrar la palabra más frecuente arriba
    plt.show()

# Generar y mostrar la "nube de palabras" para cada DataFrame (con barras)
for df_name, columns in text_columns.items():
    df = globals()[df_name]
    combined_text = get_combined_text(df, columns)
    generate_wordcloud_alt(combined_text, f'Frecuencia de Palabras para {df_name}', stopwords)
```
